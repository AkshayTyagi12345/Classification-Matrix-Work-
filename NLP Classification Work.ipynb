Write a function which accepts a CSV file location and split the data randomly in the file into 60% and 40% and save it in two different files namely “First.csv” and “Second.csv”. 
Run this function for HouseData.csv

```python
import pandas as pd 
def SplitData(location):
    df = pd.read_csv(location)
    data60per=df.sample(frac=0.6,random_state=200)
    data40per=df.drop(data60per.index)
    data60per.to_csv('First.csv')
    data40per.to_csv('Second.csv')
    print('Data Split successful')
    
SplitData("HouseData.csv")
```

Write a function which accepts a CSV and performs Label Encoding for the “Class” column of that file, creates a new Column called “Label” and saves the result in the same file.
Run this function for Marketing.csv

```python
from sklearn import preprocessing
def LabelEncoding(location):
    df=pd.read_csv(location)
    le = preprocessing.LabelEncoder()
    df['label']=le.fit_transform(df['class'])
    df.to_csv('location, index = False')
    print("Label Encoding performed ")
    
LabelEncoding('Marketing.csv')
```

Write a function which accepts a CSV which has two columns: “ActualValues” and “PredictedValues”. You need to design a confusion matrix and give values for the following:
•	Accuracy 
•	Misclassification Rate
•	True Positive Rate 
•	False Positive Rate
•	Specificity
•	Precision 
•	Null Error Rate

Also plot the ROC Curve 
Provide your understanding for each of them in a sentence or two.
Run this function for Results.csv


```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt 
```

```python
dedef PerformanceAnalysis(location):
    df = pd.read_csv(location)
    
    actualVals=df['ActualValues']
    predictVals=df['PredictedValues']
    
    print("Confusion Matrix :\n",confusion_matrix(actualVals ,predictVals))
    
    tn, fp, fn, tp = confusion_matrix(actualVals,predictVals).ravel()
    n = len(df)
    accuracy = (tp+tn)*1.0/n
    print("\nAccuracy:\n", accuracy)
    
    MCR = (fp+fn)*1.0/n
    print("\nMissclassification Rate:\n",MCR)
    
    TPR = tp*1.0/(fn+tp)
    print("\nTrue  Positive Rate:\n",TPR)
    
    FPR = fp*1.0/(tn+fp)
    print("\nFalse Positive Rate:\n",FPR)
    specificity=tn*1.0/(tn+fp)
    print("\nSpecificity:\n",specificity)
    
    precision=tp*1.0/(tp+fp)
    print ("\nPrecision:\n",precision)
    
    NER =(tp+fp)*1.0/n
    print("\nNull Error Rate:\n",NER)
    
    print("\nROC Curve:\n")
    fpr,tpr,thresold = metrics.roc_curve(actualVals,predictVals)
    roc_auc = metrics.auc(fpr,tpr)
    plt.title('Receiever operating Characteristic')
    plt.plot(fpr, tpr,'b', label = 'AUC= %0.2f' % roc_auc)
    plt.legend(loc = 'lower right')
    plt.plot([0,1],[0,1],'r--')
    plt.xlim([0,1])
    plt.ylim([0,1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
    
    
    
    
```

```python
PerformanceAnalysis('Results.csv')
```
